```
import torch
import numpy as np
import itertools
```

```
def preprocess_for_inference(sentence):
    tokens = []
    original_word_tokens = []
    for word in sentence.split(' '):
        word_tokens = tokenizer.tokenize(word)
        if len(word_tokens) > 0:
            original_word_tokens.extend([word] * len(word_tokens))
            tokens.extend(word_tokens)

    tokens += [sep_token]
    original_word_tokens += [sep_token]
    tokens = [cls_token] + tokens
    original_word_tokens = [cls_token] + original_word_tokens
    
    input_ids = tokenizer.convert_tokens_to_ids(tokens)

    # The mask has 1 for real tokens and 0 for padding tokens. Only real
    # tokens are attended to.
    attention_mask = [1] * len(input_ids)
    return input_ids, attention_mask, original_word_tokens
```

```
def preprocess_for_inference_new(sentence):
    tokens = []
    original_word_tokens = []
    for word in sentence.split(' '):
        print(word)
        word_tokens = tokenizer.tokenize(word)
        print(word_tokens)
        if len(word_tokens) > 0:
            original_word_tokens.extend([word] * len(word_tokens))
            tokens.extend(word_tokens)

    tokens += [sep_token]
    original_word_tokens += [sep_token]

    tokens = [cls_token] + tokens
    original_word_tokens = [cls_token] + original_word_tokens
    print('tokens ',tokens)
    input_ids = tokenizer.convert_tokens_to_ids(tokens)

    # The mask has 1 for real tokens and 0 for padding tokens. Only real
    # tokens are attended to.
    attention_mask = [1] * len(input_ids)
    return input_ids,attention_mask,original_word_tokens
```

```
def get_ner_prediction(input_ids, attention_mask=None):
    with torch.no_grad():
        prediction = model(input_ids, attention_mask=attention_mask)[0][0].cpu().numpy()
        prediction_after_softmax = np.exp(prediction) / np.exp(prediction).sum(-1, keepdims=True)
    return prediction_after_softmax
```

```
def processing_post_prediction(prediction, original_word_tokens, trained_entities):
    entity_names = [d['name'] for d in trained_entities if 'name' in d]
    entity_types = [d.get('type', 'Text') for d in trained_entities if 'name' in d]
    labels_idx = prediction.argmax(axis=-1)
    detected_entities = []
    answered_words = []
    for idx, label_idx in enumerate(labels_idx):
        word = original_word_tokens[idx]
        confidence = prediction[idx][label_idx].item()
        if entity_names[label_idx] != 'none' and confidence > 0.5 and word not in [cls_token,
                                                                                   sep_token] and word not in answered_words:
            answered_words.append(word)
            detected_entities += [
                {
                    "word": word,
                    "score": confidence,
                    "entity": entity_names[label_idx],
                    "type": entity_types[label_idx]
                }
            ]
    # using groupby for merging the entities with phrases
    entity_final_list = {}
    if len(detected_entities) == 0:
        return entity_final_list

    for key, group in itertools.groupby(detected_entities, lambda x: x['entity']):
        group_list = list(group)
        entity_value = ' '.join([x['word'] for x in group_list])
        average_confidence = sum([x['score'] for x in group_list]) / len(group_list)
        entity_object = {
            'confidence': average_confidence,
            'text': entity_value,
            'value': entity_value,
            'type': group_list[0]['type']
        }
        if key in entity_final_list.keys():
            entity_final_list[key].append(entity_object)
        else:
            entity_final_list[key] = [entity_object]
    return entity_final_list
```

```
cls_token="<s>"
sep_token="</s>"
text = '是的，銀行總部位於班加羅爾'
#text = 'Anirudh will visit Mumbai next week'
labels = [{'name': 'name', 'type': 'text'}, {'name': 'location', 'type': 'text'}, {'name': 'none', 'type': 'text'}]
input_ids, attention_mask, original_word_tokens = preprocess_for_inference(text.lower())
input_ids = torch.tensor(input_ids, dtype=torch.long).unsqueeze(0)
attention_mask = torch.tensor(attention_mask, dtype=torch.long).unsqueeze(0)
ner_prediction = get_ner_prediction(input_ids, attention_mask=attention_mask)
entity_prediction = processing_post_prediction(ner_prediction, original_word_tokens, labels)
print(entity_prediction)
```

```
def merge_entities(data):
    output = {}
    for i in range(len(data) - 2, -1, -1):
        if data[i]['entity'] == data[i+1]['entity']:
          if data[i]['word'][0] == '▁' and data[i+1]['word'][0] == '▁':
            data[i]['word'] = data[i]['word'] + ' ' + data[i+1]['word']
          else:
            data[i]['word'] = data[i]['word']  + data[i+1]['word']          
          data.pop(i + 1)
    for word in data:
      word['word'] = word['word'].replace('▁',' ').strip()
      output[word['entity']] = [{'confidence':word['score'],'text':word['word'],'value':word['word'],'type': 'text'}]
    return output
```
